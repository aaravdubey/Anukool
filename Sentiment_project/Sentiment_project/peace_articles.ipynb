```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
from selenium import webdriver
browser=webdriver.Chrome()
browser.get("https://www.innerpeacenow.com/inner-peace-blog")
```

```python
soup=BeautifulSoup(browser.page_source,'html.parser')
print(soup)
```

```python
dj=[]
links=[]
s=(soup.find('div',class_="blog-list"))
for sp in s.find_all('article'):
    title=sp.find('h1').text[1:-1]
    cont=''
    link=sp.find('p',class_="entry-more-link")
    final_link=link.find('a').get('href')
    paragraph=sp.find_all('p',class_="")
    for p in paragraph:
        cont=cont+p.text
    dj.append([title,cont])
    links.append(final_link)
print(len(dj))
```

```python
df=pd.DataFrame(dj,columns=["Heading ","Context"])
df
```

```python
links_data=[]
links
browser=webdriver.Chrome()
for link in links:
    browser.get("https://www.innerpeacenow.com"+link)
    soup=BeautifulSoup(browser.page_source,'html.parser')
    cont=''
    sp=soup.find('div',class_='col sqs-col-12 span-12')
    paragraph=sp.find_all('p')
    for p in paragraph:
        cont=cont+p.text
    links_data.append(cont)
print(len(links_data))
```

```python
df['Content']=links_data
df
```
